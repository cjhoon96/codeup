```python
import sys
input=sys.stdin.readline

m,n = map(int,input().split())

on_graph = []
off_graph = []
for _ in range(m):
    a = list(map(int,input().split()))
    b = a.copy()
    on_graph.append(a)
    off_graph.append(b)
    
def turn_on(x,y):
    if 0<=x<n and 0<=y<m and on_graph[y][x] == 0:
        on_graph[y][x] = 1
        turn_on(x+1,y)
        turn_on(x-1,y)
        turn_on(x,y+1)
        turn_on(x,y-1)
        return True
    return False
    
def turn_off(x,y):
    if 0<=x<n and 0<=y<m and off_graph[y][x] == 1:
        off_graph[y][x] = 0
        turn_off(x+1,y)
        turn_off(x-1,y)
        turn_off(x,y+1)
        turn_off(x,y-1)
        return True
    return False

on_cnt = 0
off_cnt = 0

for y in range(m):
    for x in range(n):
        if turn_on(x,y):
            on_cnt += 1

for y in range(m):
    for x in range(n):
        if turn_off(x,y):
            off_cnt += 1  

print(on_cnt, off_cnt)
```
