```python
omok = []
for _ in range(19):
    omok.append(list(map(int,input().split())))


def hori_cnt(color,y,x):
    global hor_cnt
    if omok[y][x] == color and 0<=y<=18 and 0<=x<=18:
        hor_cnt += 1
        if x == 18:
            return
        else:
            hori_cnt(color,y,x+1)
    else:
        return
    
def verti_cnt(color,y,x):
    global ver_cnt
    if omok[y][x] == color and 0<=y<=18 and 0<=x<=18:
        ver_cnt += 1
        if y == 18:
            return
        else:
            verti_cnt(color,y+1,x)
    else:
        return
    
def R_diag_cnt(color,y,x):
    global R_di_cnt
    if omok[y][x] == color and 0<=y<=18 and 0<=x<=18:
        R_di_cnt += 1
        if x == 18 or y == 18:
            return
        else:
            R_diag_cnt(color,y+1,x+1)
    else:
        return

def L_diag_cnt(color,y,x):
    global L_di_cnt
    if omok[y][x] == color and 0<=y<=18 and 0<=x<=18:
        L_di_cnt += 1
        if x == 0 or y == 18:
            return
        else:
            L_diag_cnt(color,y+1,x-1)
    else:
        return

def last_check(color,a,x,y):
    if a == 0:
        if x == 0 or omok[y][x-1] != color:
            return True
        else:
            return False
    elif a == 1:
        if y == 0 or omok[y-1][x] != color:
            return True
        else:
            return False
    elif a == 2:
        if y == 0 or x == 18 or omok[y-1][x+1] != color:
            return True
        else:
            return False
    elif a == 3:
        if y == 0 or x == 0 or omok[y-1][x-1] != color:
            return True
        else:
            return False
    
for y in range(19):
    for x in range(19):
        
        hor_cnt, ver_cnt, L_di_cnt, R_di_cnt = 0, 0, 0, 0
        
        if omok[y][x] in (1,2):
            color = omok[y][x]
            hori_cnt(color,y,x)
            verti_cnt(color,y,x)
            L_diag_cnt(color,y,x)
            R_diag_cnt(color,y,x)
        
        cnt_list = (hor_cnt, ver_cnt, L_di_cnt, R_di_cnt)
        
        check = False
        if 5 in cnt_list:
            lst = []
            for a in range(4):
                if cnt_list[a] == 5:
                   check = last_check(color,a,x,y)
        if check:
            break
    if check:
        break

else:
    print(0)

if hor_cnt == 5 or ver_cnt == 5 or R_di_cnt == 5:
    print(color)
    print(y+1, x+1)
elif L_di_cnt == 5:
    print(color)
    print(y+5, x-3)
```
