문제 설명    
영일이는 숫자 0에서 n개의 수를 덧셈과 뺄셈을 이용하여 m으로 만들려고 한다.

 

예를 들어 n=2이고 1, -1로 m=0을 만드는 경우는 다음과 같다.

 

0 + 1 + (-1) = 0

0 - 1 - (-1) = 0

 

n개의 수를 덧셈과 뺄셈을 이용하여 m이 되는 경우의 수를 모두 구하시오.

입력
첫 번째 줄에 m이 입력된다.

두 번째 줄에 n이 입력된다. (0<=n<=20)

세 번째 줄에 n개의 그 수들이 입력된다.  (입력되는 수들의 최대 합은 int범위 이내이다.)

출력
m을 만들 수 있는 경우의 수를 출력한다. 입력되는 수들은 모두 사용해야 하며 한 번씩만 사용할 수 있다.

입력 예시   
10
3
1 4 5

출력 예시
1

도움말
1 4 5로 10을 만들 수 있는  방법은,

1+4+5=10 밖에 없으므로 1을 출력한다.

 

```python
from itertools import product

m = int(input()) # 목표 숫자
n = int(input()) # 입력받는 숫자 개수

cnt = 0 # 카운트용 변수

if n != 0:
    data = list(map(int, input().split())) # 주어진 n개의 숫자들
else:
    data = []

data.sort(reverse=True)

minus_dic={}
jdx=1
for i in product([1,-1], repeat=n):
    minus_dic[jdx]=i
    jdx+=1

k=1
while k<=2**n:      # 전부다 확인 하는 척
    now_sum=0
    minu=minus_dic[k]
    
    for idx in range(n):
        now_sum+=data[idx]*minu[idx]  # 인덱스 idx까지의 데이터 계산 값
        later_sum=sum(data[idx+1:])   # 인덱스 idx 이후의 데이터를 모두 더한 값
        
        if ((now_sum<m and now_sum+later_sum<m) or (now_sum>m and now_sum-later_sum>m)): #더이상 볼 필요가 없으면

            while k<=2**n and minus_dic[k][idx]==minus_dic[k][idx]:
                    k+=1
            break
    else:
        if now_sum==m:
            cnt+=1
            k+=1
print(cnt)
```
