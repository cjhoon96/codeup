```python
import sys
input=sys.stdin.readline
from collections import deque

n, k = map(int,input().split())

bina_code = ['none']
for _ in range(n):
    bina_code.append(input())
    
a,b = map(int,input().split())

def ham(i,j):
    cnt = 0
    for idx in range(k):
        if bina_code[i][idx] != bina_code[j][idx]:
            cnt += 1
        if cnt > 1:
            return False
    else:
        return True

dis_1 = {}                                     ## 입력 받은 이진 코드들을 bina_code에서의 인덱스값으로 변환하고 
for check_1 in range(1,n+1):
    for check_2 in range(check_1+1,n+1):
        if ham(check_1,check_2):                     ## 해밍 거리가 서로 1이면 각 지점들을 키값으로 밸류값은 인접한 지점들의 리스트로 저장한 인접그래프를 만든다.
            if check_1 in dis_1:
                dis_1[check_1].append(check_2)        
            else:
                dis_1[check_1] = [check_2]
            if check_2 in dis_1:
                dis_1[check_2].append(check_1)
            else:
                dis_1[check_2] = [check_1]

if b in dis_1[a]: ## 만일 a와 b의 해밍거리가 1이라면 a b를 출력한다.
    print(a,b)

else: ## 그렇지 않다면 bfs를 통해 최단 경로를 찾는다.
    check_vs = [i for i in range(1,n+1)]  ##아직 방문하지 않은 코드번호가 저장될 리스트
    hamming_dis = {1:[]} ## 키:a와의 거리 밸류:[(직전에 방문한 코드 번호 , 지금 방문한 코드번호), ... ] 이후 b가 나왔을시 역으로 찾아가며 온 경로를 추적하기 위해!!
    check_vs.remove(a)  ## a는 미리 방문 처리를 한다
    
    for d in dis_1[a]:  ## a와 거리가 1이었던 코드 번호를 미리 방문처리하고 hamming_dis의 1 키값에 저장한다.
        check_vs.remove(d)
        hamming_dis[1].append((a,d))
        
    queue = deque(dis_1[a]) ## 위에서 저장했던 거리 1인 코드 번호들을 큐에 저장한다.
    
    b_check = False ## 이후 결과에서 b까지 도달 했는지를 체크  (b에 도달 하였다면 True로 바꿔준다.)
    while check_vs and queue: ## 더이상 인접한 코드 번호들 중에 방문하지 않은 코드가 없어서 queue가 비거나 전부다 방문한 경우에 와일문을 탈출한다. 
        chck = queue.popleft()  ## 큐에서 하나를 추출한다.
        
        for now_dis in range(1,len(hamming_dis)+1): ## 추출한 코드 번호가 a로 부터 몇번째 거리에 있는지
            for code in hamming_dis[now_dis]:
                if chck == code[1]:
                    break
            if chck == code[1]:
                break
        
        for j in dis_1[chck]:      
            if j in check_vs and j not in queue:
                check_vs.remove(j)
                queue.append(j)
                now_dis +=1
                if now_dis in hamming_dis:
                    hamming_dis[now_dis].append((chck,j))
                else:
                    hamming_dis[now_dis] = [(chck,j)]
            if j == b:
                b_check = True
                break
        if j == b:
            break

    if b_check:
        print_list = [b]
        rev_check = b
        while now_dis > 0:
            for i in hamming_dis[now_dis]:
                if i[1] == rev_check:
                    break
            rev_check = i[0]
            print_list.append(rev_check)
            now_dis -= 1
        print_list.reverse()
        for i in print_list:
            print(i, end=' ')
    
    else:
        print(-1)
                
```
