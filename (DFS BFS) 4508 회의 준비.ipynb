```python
import sys
input=sys.stdin.readline
from collections import deque

n = int(input())
m = int(input())

fr_list = [[] for _ in range(n+1)]

for _ in range(m):
    a,b = map(int,input().split())
    fr_list[a].append(b)
    fr_list[b].append(a)

visit_list = [1 for _ in range(n+1)]


def committee(a):
    group.append(a)
    visit_list[a] = 0
    queue = deque(fr_list[a])
    while queue:
        check = queue.popleft()
        if visit_list[check]:
            group.append(check)
            visit_list[check] = 0
            for i in fr_list[check]:
                if visit_list[i] and i not in queue:
                    queue.append(i)

groups = []

for p in range(1,n+1):
    if visit_list[p]:
        group = []
        committee(p)
        groups.append(group)

print(len(groups))


print_list = []


for gr in groups:
    vote = []
    for idx in gr:
        check_gr = gr.copy()

        check_gr.remove(idx)
        
        depth = { 1:fr_list[idx] }
        
        queue = deque(fr_list[idx])

        for c in fr_list[idx]:
            check_gr.remove(c)

        while check_gr and queue:
            chck = queue.popleft()

            for d in range(1,len(depth)+1):
                if chck in depth[d]:
                    break
            
            for j in fr_list[chck]:      
                if j in check_gr and j not in queue:
                    check_gr.remove(j)
                    queue.append(j)
                    if d+1 in depth:
                        depth[d+1].append(j)
                    else:
                        depth[d+1] = [j]

        vote.append([len(depth),idx])
    vote.sort()

    print_list.append(vote[0][1])
print_list.sort()
for i in print_list:
    print(i)
    
```
