```python
import sys
input=sys.stdin.readline
from collections import deque

m,n,h = map(int,input().split())

tmt = []
for _ in range(h):
    height = []
    for _ in range(n):
        height.append(list(map(int,input().split())))
    tmt.append(height)

queue = deque()
cnt = 0

for z in range(h):
    for y in range(n):
        for x in range(m):
            check = tmt[z][y][x]
            if check == 1:
                cnt += 1
                queue.append([z,y,x])
            elif check == -1:
                cnt += 1

max_day = 0

while queue:
    t = queue.popleft()
    now_day = tmt[t[0]][t[1]][t[2]]+1
    if t[0]-1 >= 0 and tmt[t[0]-1][t[1]][t[2]] == 0:
        tmt[t[0]-1][t[1]][t[2]] = now_day
        cnt += 1
        queue.append([t[0]-1,t[1],t[2]])
        if now_day+1 > max_day:
            max_day = now_day
    if t[0]+1 <h and tmt[t[0]+1][t[1]][t[2]] == 0:
        tmt[t[0]+1][t[1]][t[2]] = now_day
        cnt += 1
        queue.append([t[0]+1,t[1],t[2]])
        if now_day > max_day:
            max_day = now_day
    if t[1]-1 >= 0 and tmt[t[0]][t[1]-1][t[2]] == 0:
        tmt[t[0]][t[1]-1][t[2]] = now_day
        cnt += 1
        queue.append([t[0],t[1]-1,t[2]])
        if now_day > max_day:
            max_day = now_day
    if t[1]+1 <n and tmt[t[0]][t[1]+1][t[2]] == 0:
        tmt[t[0]][t[1]+1][t[2]] = now_day
        cnt += 1
        queue.append([t[0],t[1]+1,t[2]])
        if now_day > max_day:
            max_day = now_day
    if t[2]-1 >= 0 and tmt[t[0]][t[1]][t[2]-1] == 0:
        tmt[t[0]][t[1]][t[2]-1] = now_day
        cnt += 1
        queue.append([t[0],t[1],t[2]-1])
        if now_day > max_day:
            max_day = now_day
    if t[2]+1 <m and tmt[t[0]][t[1]][t[2]+1] == 0:
        tmt[t[0]][t[1]][t[2]+1] = now_day
        cnt += 1
        queue.append([t[0],t[1],t[2]+1])
        if now_day > max_day:
            max_day = now_day

if cnt == m*n*h:
    print(max_day-1)
else:
    print(-1)
```
